**BINNING**

```python
#Binning
df['YearBuiltBins'] = pd.cut(df['yr_built'], bins=5, labels=False)
df['SqftAboveBins'] = pd.qcut(df['sqft_above'], q=5, labels=False)
df
```

```python
# Histogram of "YearBuiltBins"
plt.hist(df['YearBuiltBins'])
plt.xlabel('Year Built Bins')
plt.ylabel('Count')
plt.title('Distribution of Year Built')
plt.show()
```

**NORMALIZATION**

```python
numerical_cols = ['price', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors', 'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated']
df[numerical_cols] = (df[numerical_cols] - df[numerical_cols].min()) / (df[numerical_cols].max() - df[numerical_cols].min())
df
```

**Z-score Normalization**

```python

df[numerical_cols] = (df[numerical_cols] - df[numerical_cols].mean()) / df[numerical_cols].std()
df
```

```python

# Example visualization: Scatter plot of "SqftAboveBins" vs "price"
plt.scatter(df['SqftAboveBins'], df['price'])
plt.xlabel('Sqft Above Bins')
plt.ylabel('Price')
plt.title('Scatter plot of Sqft Above Bins vs Price')
plt.show()
```
